```python
from nltk import word_tokenize
from collections import Counter
from konlpy.tag import Kkma
```

```python
f = open("삼포 가는 길.txt", "r",encoding='utf-8')

data = f.read()
```

```python
import nltk
nltk.download('punkt')
```

```python
kkma = Kkma()
wordss=kkma.pos(data)
nouns = [n for n, tag in wordss if tag in ["NNG","NNP","NNB"] ]
print(nouns)
```

```python
stopwords = ['것','거','뿐','수','거','곳','참','데','때','아','라도','하면','서',"추"]
result = []

for w in nouns:
  if w not in stopwords:
    #print(w)
    result.append(w)
print(result)
```

```python
vocab=Counter(result)
print(vocab)
```

```python
import operator 

for k, v in vocab.items():
  if v >= 5:
    print(k,v)
```
